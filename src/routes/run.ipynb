```python
import json
```

```python

with open('./eng_list.json') as f:
    data = json.load(f)

```

```python
repo = {}

for d in data:
    l = len(d)

    try:
        repo[l].append(d)
    except:
        repo[l] = [d]
```

```python
keys = list(repo.keys())
keys.sort()
keys
```

```python
num = [len(repo[k]) for k in keys]
num
```

```python
import matplotlib.pyplot as plt

plt.plot(keys, num)
```
